# 5초씩 모든 파일들을 각각 폴더에 자르는 코드

```python
# 설명서
# 몇초씩 끊을지 정하려면 코드에서 duration=  숫자를 변경하면 가능
```

```python
import subprocess
import os


def cut_save_mp4(input_file, output_file, start_time, end_time):
    command = [
        'ffmpeg',
        '-i', input_file,
        '-ss', start_time,
        '-to', end_time,
        '-c', 'copy',
        output_file
    ]
    subprocess.run(command)

# 여기서 duration=  :  몇초씩 영상을 자를지 선택가능
def cut_save_mp4_increments(input_file, output_folder, output_base_name, duration=5):
    command = [
        'ffprobe',
        '-v', 'error',
        '-show_entries', 'format=duration',
        '-of', 'default=noprint_wrappers=1:nokey=1',
        input_file
    ]
    result = subprocess.run(command, capture_output=True, text=True)
    total_duration = float(result.stdout)
    start_time = 0
    end_time = duration

    while start_time < total_duration:
        output_file = os.path.join(output_folder, f"{output_base_name}{start_time}{end_time}.mp4")
        cut_save_mp4(input_file, output_file, format_time(start_time), format_time(end_time))
        start_time += duration
        end_time += duration
        end_time = min(end_time, total_duration)

def format_time(time):
    hours = int(time / 3600)
    minutes = int((time % 3600) / 60)
    seconds = int(time % 60)
    return f"{hours:02d}:{minutes:02d}:{seconds:02d}"

def process_videos(input_folder, output_folder):
    video_files = os.listdir(input_folder)
    os.makedirs(output_folder, exist_ok=True)

    for file_name in video_files:
        if file_name.endswith('.mp4'):
            input_file = os.path.join(input_folder, file_name)
            output_base_name = os.path.splitext(file_name)[0]
            output_subfolder = os.path.join(output_folder, output_base_name)
            os.makedirs(output_subfolder, exist_ok=True)
            cut_save_mp4_increments(input_file, output_subfolder, output_base_name)

# input폴더에 영상이 들어있는 폴더 지정
# output폴더는 5초씩 자른 최종 영상들을 저장할 폴더 지정
# output_folder 내에 동일한 이름을 가진 하위 폴더를 생성하고
# 그 안에 자른 동영상 파일들이 저장됩니다. 
input_folder = r'/Users/hwangjong-won/Desktop/DL'
output_folder = r'/Users/hwangjong-won/Desktop/DL/testset'
process_videos(input_folder, output_folder)
```
